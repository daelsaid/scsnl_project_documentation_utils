#### 6. Set 4 empty lists (these will be populated as each subject's template is checked)

#### 8. If each key string is not found in the import formula tab, print 'assessor/score1/score2 complete/no checks needed for PID, visit #, assessment name'

````python
needs_assessor=[]
needs_score1=[]
needs_score2=[]
needs_check=[]


for subj_data_file in os.listdir(all_assess_dirs):
    file_prefix=subj_data_file.split('.')[0]
    try:
        subject=file_prefix.split('_')[-2]
        visit=file_prefix.split('_')[-1]
        assessment=file_prefix.split('_9')[0].split('MET_')[1]
        subj_df=pd.read_excel(subj_data_file,sheetname='Import_Formula',header=[0],skiprows=None,dtype=str)

        if (subj_df.values=='Assessor to complete').any() == True:
            needs_assessor.append([subject,visit,assessment])
        else:
            print subject,visit,assessment,': Assessor scoring complete'
        if (subj_df.values=='Scorer1 to complete').any() == True:
            needs_score1.append([subject,visit,assessment])
        else:
            print subject,visit,assessment,': Score1 scoring complete'
        if (subj_df.values=='Scorer2 to complete').any() == True:
            needs_score2.append([subject,visit,assessment])
        else:
            print subject,visit,assessment,': Scorer2 scoring complete'
        if (subj_df.values=='check').any() == True:
            needs_check.append([subject,visit,assessment])
    except:
        print ""
        print('error in cell values for:',file_prefix,'file cannot be read')
        continue
````
